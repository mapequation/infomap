# 5.3 Modeling temporal data

This notebook accompanies Section 5.3 of [*Community Detection with the Map Equation and Infomap: Theory and Applications*](https://doi.org/10.1145/3779648).

You can represent temporal data in several ways. When you have network snapshots taken over different time intervals, a temporal multilayer network organizes the snapshots into one layer per time point or interval, which lets you trace how communities evolve over time.

## Using a multilayer network

In a multilayer network, each physical node can appear in several layers, with a different link structure in each one. You may declare the physical nodes explicitly, as in the Pajek format, but you have three ways to specify the links, depending on the data you have. Under the `*Multilayer` heading, you list every intra-layer and inter-layer link explicitly. Under the `*Intra` heading, you list only the links within each layer and let inter-layer relaxation generate the transitions between layers, optionally constrained by inter-layer links you declare under the `*Inter` heading.

See the [interactive storyboard](https://mapequation.org/apps/multilayer-network/index.html) for an illustration.

## Temporal constraints

When you let Infomap generate the inter-layer links, a node in one layer can connect to nodes in any other layer. To control temporal ordering, use `--multilayer-relax-limit` to restrict the random walker to layers within a given distance in each direction. Rojas et al. set `--multilayer-relax-limit 1` in ["A multiscale view of the Phanerozoic fossil record reveals the three major biotic transitions"](https://www.nature.com/articles/s42003-021-01805-y) to model a time-reversible causal simulation connecting geological stages.

By default, a node couples to its counterpart in other layers with a strength proportional to its outgoing weight. With the `--multilayer-relax-by-jsd` flag, Infomap instead couples layers more strongly where a node's connectivity patterns are similar, measured by the Jensen-Shannon divergence. Aslak et al. use this approach in ["Constrained information flows in temporal networks reveal intermittent communities"](https://journals.aps.org/pre/abstract/10.1103/PhysRevE.97.062312) to identify intermittent communities that form and dissolve over time.
