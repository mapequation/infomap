# 5.2 Multilayer networks

This notebook accompanies Section 5.2 of [*Community Detection with the Map Equation and Infomap: Theory and Applications*](https://doi.org/10.1145/3779648).

Multilayer flows are stratified across layers: people commute between cities by different modes of transport, or message each other on different platforms. Each physical node lives in many layers, and Infomap treats the result as a memory network whose state nodes correspond to a physical node in a given layer. Intra-layer links connect state nodes within a layer; inter-layer links connect them across layers. How you supply those inter-layer links sets the input format:

- Explicit multilayer links: inter-layer link weights are given directly
- Intra- and inter-layer links: intra-layer links plus coupling weights between a physical node's layers
- Intra-layer links and a relax rate $r$: no inter-layer links, the walker switches layers with probability $r$

```python
from infomap import Infomap
```

## Explicit multilayer links

When the data give the inter-layer link weight $w_{u^A v^B}$ between node $u$ in layer $A$ and node $v$ in layer $B$ directly, add each link with `add_multilayer_link`.

```python
im = Infomap(silent=True, directed=True, seed=123)

# intra
# (source_layer, node), (target_layer, node), weight
im.add_multilayer_link((1, 1), (1, 4), 0.8)
im.add_multilayer_link((1, 4), (1, 1), 1)
im.add_multilayer_link((1, 1), (1, 5), 0.8)
im.add_multilayer_link((1, 5), (1, 1), 1)
im.add_multilayer_link((1, 4), (1, 5), 1)
im.add_multilayer_link((1, 5), (1, 4), 1)
im.add_multilayer_link((2, 1), (2, 2), 0.8)
im.add_multilayer_link((2, 2), (2, 1), 1)
im.add_multilayer_link((2, 1), (2, 3), 0.8)
im.add_multilayer_link((2, 3), (2, 1), 1)
im.add_multilayer_link((2, 2), (2, 3), 1)
im.add_multilayer_link((2, 3), (2, 2), 1)
# inter
im.add_multilayer_link((1, 1), (2, 2), 0.2)
im.add_multilayer_link((1, 1), (2, 3), 0.2)
im.add_multilayer_link((2, 1), (1, 4), 0.2)
im.add_multilayer_link((2, 1), (1, 5), 0.2)

im.run()
print(f"Found {im.num_top_modules} modules with codelength {im.codelength:.2f} bits")
```

```python
# requires pandas
im.get_dataframe()
```

## Intra- and inter-layer links

When the data give intra-layer links and inter-layer coupling weights $\upsilon_{u^A u^B}$ separately, add them with `add_multilayer_intra_link` and `add_multilayer_inter_link`. Infomap expands the coupling into multilayer links.

```python
im = Infomap(silent=True, directed=True, seed=123)

# intra
# layer, source_node, target_node, weight
im.add_multilayer_intra_link(1, 1, 4, 0.8)
im.add_multilayer_intra_link(1, 4, 1, 1)
im.add_multilayer_intra_link(1, 1, 5, 0.8)
im.add_multilayer_intra_link(1, 5, 1, 1)
im.add_multilayer_intra_link(1, 4, 5, 1)
im.add_multilayer_intra_link(1, 5, 4, 1)
im.add_multilayer_intra_link(2, 1, 2, 0.8)
im.add_multilayer_intra_link(2, 2, 1, 1)
im.add_multilayer_intra_link(2, 1, 3, 0.8)
im.add_multilayer_intra_link(2, 3, 1, 1)
im.add_multilayer_intra_link(2, 2, 3, 1)
im.add_multilayer_intra_link(2, 3, 2, 1)
# inter
# source_layer, node, target_layer, weight
im.add_multilayer_inter_link(1, 1, 2, 0.2)
im.add_multilayer_inter_link(2, 1, 1, 0.2)

im.run()
print(f"Found {im.num_top_modules} modules with codelength {im.codelength:.2f} bits")
```

```python
im.get_dataframe()
```

## Intra-layer links and a relax rate

When no inter-layer links are available, set a relax rate $r$ instead. The walker follows intra-layer links with probability $1-r$ and switches layers with probability $r$, so only intra-layer links go in.

```python
im = Infomap(silent=True, directed=True, seed=123, multilayer_relax_rate=0.15)

# intra
# layer, source_node, target_node, weight
im.add_multilayer_intra_link(1, 1, 4, 0.8)
im.add_multilayer_intra_link(1, 4, 1, 1)
im.add_multilayer_intra_link(1, 1, 5, 0.8)
im.add_multilayer_intra_link(1, 5, 1, 1)
im.add_multilayer_intra_link(1, 4, 5, 1)
im.add_multilayer_intra_link(1, 5, 4, 1)
im.add_multilayer_intra_link(2, 1, 2, 0.8)
im.add_multilayer_intra_link(2, 2, 1, 1)
im.add_multilayer_intra_link(2, 1, 3, 0.8)
im.add_multilayer_intra_link(2, 3, 1, 1)
im.add_multilayer_intra_link(2, 2, 3, 1)
im.add_multilayer_intra_link(2, 3, 2, 1)

im.run()
print(f"Found {im.num_top_modules} modules with codelength {im.codelength:.2f} bits")
```

```python
im.get_dataframe()
```
